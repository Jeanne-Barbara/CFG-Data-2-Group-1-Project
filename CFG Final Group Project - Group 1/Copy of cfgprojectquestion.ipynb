```python
#
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("olympicdatatousex.csv")
df

```

```python
newdata1= df[['Team', 'Year', 'Event']].copy()
newdata1
```

```python
Finalnewdata= newdata1.groupby(['Team','Year'])['Event'].count().reset_index(name="Eventcount")
Finalnewdata
```

```python
Yearsort= Finalnewdata.sort_values(by=['Year'], ascending=True)
Yearsort
```

```python
Eventsort= Yearsort.sort_values(by=['Eventcount'], ascending=False)
Eventsort
```

```python
import numpy as np
import matplotlib.pyplot as plt

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
Year1 = [744, 708, 592, 559, 547, 0]
Year2 = [669, 0, 541, 506, 494, 672]
Year3 = [699, 483, 0, 510, 528, 470]

# Set position of bar on X axis
br1 = np.arange(len(Year1))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, Year1, color ='blue', width = barWidth,
		edgecolor ='grey', label ='2008')
plt.bar(br2, Year2, color ='orange', width = barWidth,
		edgecolor ='grey', label ='2012')
plt.bar(br3, Year3, color ='purple', width = barWidth,
		edgecolor ='grey', label ='2016')

# Adding Xticks
plt.xlabel('Team', fontweight ='bold', fontsize = 15)
plt.ylabel('Eventcount', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Year1))],
		['USA', 'CHN', 'RUS', 'AUS', 'GER', 'GBR'])
plt.title('How many events the top 6 Teams have participated in accross 3 Olympics')

plt.legend()
plt.show()

```
